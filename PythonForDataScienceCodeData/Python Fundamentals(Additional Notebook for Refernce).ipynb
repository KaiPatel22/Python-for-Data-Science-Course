<font color=Navy><b>two types of Errors in Python :<br>
1) Syntax errors<br>

exceptions are raised when some internal events occur which changes the normal flow of the program.

Exceptions:
Exceptions are raised when the program is syntactically correct but the code resulted in an error.
This error does not stop the execution of the program, however, it changes the normal flow of the program.

```python
a,b=30,0

if (a/b) == 0:
    print('yes')
else:
    print('no')
```

<font color=Navy><b>   In the above code exception is raised

Exception handling using : Try and except

Please visit this website to know morw about different exceptions :https://docs.python.org/2/library/exceptions.html#exception-hierarchy

```python
a,b=30,0
try:  
    if (a/b) == 0:
        print('yes')
    else:
        print('no') 
  
except ZeroDivisionError: 
    print("An error occurred") 
```
